**–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü –≥–æ–¥ `year`, –≤ –∫–æ—Ç–æ—Ä—ã–π –±—ã–ª–∞ –≤—ã—Å—Ç–∞–≤–ª–µ–Ω–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏—è**

```python
data['year'] = data['first_day_exposition'].dt.year
```

**–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Ç–∏–ø —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã `type_floor`, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞–∂–µ—Ç `–ø–µ—Ä–≤—ã–π`, `–ø–æ—Å–ª–µ–¥–Ω–∏–π`, `–¥—Ä—É–≥–æ–π`**

```python
def type_floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    if 1 < floor == floors_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–¥—Ä—É–≥–æ–π'
```

```python
data['type_floor'] = data.apply(type_floor, axis=1)
```
